# **Bivariate Analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
titanic= pd.read_csv('../files/train.csv')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

**1. Scatterplot(Num - Num)**

```python
plt.figure(
    figsize=(7, 7),
    dpi=150,
)
sns.scatterplot(data=flights, x="year", y="passengers", hue="month", size="passengers")
```

**2. Bar Plot(Num - Categorical)**

```python
sns.barplot(x=titanic["Pclass"], y=titanic["Age"], hue=titanic['Sex'])

# this gives average
```

**3. Box Plot(Numerical - Categorical)**

```python
sns.boxplot(x = titanic['Sex'], y = titanic['Age'], hue=titanic['Survived'])
```

**4. Displot(Numerical - Categorical)**

```python
sns.displot(
    titanic[titanic["Survived"] == 0]["Age"],
    fill=True,
    kind="kde",
)
sns.displot(
    titanic[titanic["Survived"] == 1]["Age"],
    kind="kde",
    fill=False,
)
```

**5. HeatMap(Categorical - Categorical)**

```python
df = pd.crosstab(titanic['Pclass'], titanic['Survived'])
df
```

```python
sns.heatmap(data=df, annot=True)
```

**6. ClusterMap(Categorical - Categorical)**

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

**7. PairPlot**

```python
sns.pairplot(iris, hue='species')
```

```python
data_flight = flights.pivot_table(columns="year", index="month", values="passengers")
sns.heatmap(data_flight)
```
