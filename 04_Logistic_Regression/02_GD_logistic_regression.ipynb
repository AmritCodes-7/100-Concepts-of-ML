# Logistic Regression from Scratch


```python
from sklearn.datasets import make_classification
import numpy as np
```

```python
X, y = make_classification(
    n_samples=200,
    n_features=2,
    n_informative=1,
    n_redundant=0,
    n_classes=2,
    n_clusters_per_class=1,
    random_state=50,
    hypercube=False,
    class_sep=20,
)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=[10, 6])
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=100)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(solver="sag", penalty=None)
lr.fit(X, y)
```

```python
print(f"{lr.intercept_} {lr.coef_}")
```

```python
def sigmoid(z):
    return 1 / (1 - np.exp(-z))
```

```python
def gd(X, y, epochs, lr):
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    learning_rate = lr

    for _ in range(epochs):
        y_hat = sigmoid(np.dot(X, weights))
        weights = weights + (learning_rate / X.shape[0]) * (np.dot(y - y_hat, X))
    return weights[1:], weights[0]
```

```python
coef_, intercept_ = gd(X, y, 5000,0.1)
```

```python
print(f"{intercept_} {coef_}")
```
