**Simple Linear Regression**

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../files/placement.csv")
```

```python
data.sample(5)
```

```python
import matplotlib.pyplot as plt
plt.scatter(data['cgpa'], data['package'])
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
X = data.iloc[:, 0:1]
y = data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=54
)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression()
```

```python
slr.fit(X_train, y_train)
```

```python
y_pred = slr.predict(X_test)
```

```python
plt.scatter(data["cgpa"], data["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
plt.plot(X_train, slr.predict(X_train), color="red") # is the line used by SLR
```

```python
# y= mx + b
m = slr.coef_
b = slr.intercept_
print(m,b)
```
